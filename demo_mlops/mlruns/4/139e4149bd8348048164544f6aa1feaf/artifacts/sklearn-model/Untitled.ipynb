```python
artifacts = {
    "sklearn_model": sklearn_model_path
}
```

```python
from pydantic import BaseModel
```

```python
class InputFeatures(BaseModel):
    fixed_acidity: float 
    volatile_acidity: float 
    citric_acid: float 
    residual_sugar: float
    chlorides: float
    free_sulfur_dioxide: float
    total_sulfur_dioxide: float
    density: float
    pH: float
    sulphates: float
    alcohol: float
```

```python
class SKLearnWrapper(mlflow.pyfunc.PythonModel):
    def load_context(self, context):
        import pickle
        self.sklearn_model = pickle.load(open(context.artifacts["sklearn_model"], 'rb'))
    def predict(self, model, InputFeatures):
        return self.sklearn_model.predict(InputFeatures)
```

```python
mlflow_pyfunc_model_path = "sklearn_mlflow_pyfunc_custom3"
```

```python
mlflow.pyfunc.save_model(path=mlflow_pyfunc_model_path, python_model=SKLearnWrapper(), conda_env=conda_env, artifacts=artifacts)
```
